```python
import pandas as pd
import os

FigueData = pd.read_excel(os.path.join("graphdata","dataFigueroa.xlsx"),
                          index_col=0)
```

```python
FigueData.head()
```

```python
FigueData.columns
```

```python
varsToDrop=['Multinacional']
adjacency=FigueData.drop(varsToDrop,axis=1)

# Resultado:
adjacency
```

```python
import networkx as nx

EliteNet = nx.from_pandas_adjacency(adjacency)
```

```python
# nodes:
len(EliteNet)

# edges:
EliteNet.size()

# plot
nx.draw_random(EliteNet,
                node_color='red',
                edge_color='lightblue',
                with_labels=True,
                font_size=8)
```

```python
EliteNet.remove_edges_from(nx.selfloop_edges(EliteNet))
```

```python
nx.draw_random(EliteNet,
                node_color='red',
                edge_color='lightblue',
                with_labels=True,
                font_size=8)
```

```python
FigueData['Multinacional'].head()
```

```python
# no attribute:
EliteNet.nodes.data()
```

```python
dict(zip(FigueData.index,FigueData['Multinacional']))
```

```python
attributeToAdd=dict(zip(FigueData.index,FigueData['Multinacional']))
nx.set_node_attributes(EliteNet, attributeToAdd,'multi')
```

```python
colors={e:'pink' if v==1 else 'red' for e,v in nx.get_node_attributes(EliteNet,"multi").items()}
nx.set_node_attributes(EliteNet, colors, "color")

# updated nodes
EliteNet.nodes.data()
```

```python
# using node attributes

nx.draw_random(EliteNet)
```

```python
from pyvis.network import Network
import networkx as nx
import pandas as pd
from pyvis.network import Network
from IPython.display import display, HTML

Gvis = Network('500px', '500px', notebook=True, cdn_resources='remote')
Gvis.from_nx(EliteNet.copy())
Gvis.show('Gvis.html')
```

```python
for node in EliteNet.nodes(data=True):
    Gvis.get_node(node[0])['color'] = node[1]['color']
```

```python
with open('Gvis.html', 'r') as f:
    html_content = f.read()
display(HTML(html_content))
```
